```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def imread_rgb(path):
    img_bgr = cv2.imread(path, cv2.IMREAD_COLOR)
    return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

def imread_gray(path):
    img_bgr = cv2.imread(path)
    img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    return img_gray
```

```python
array = np.array([[1, 0],
                  [0, 1]])
plt.imshow(array, cmap='gray')
```

```python
array_spectrum = np.array([np.arange(0,255,15),
                           np.arange(255,0,-15)])
plt.imshow(array_spectrum, cmap='gray')
```

```python
sunshine = imread_rgb('sunshine.jpg')
plt.imshow(sunshine)
print(f"Image shape: {sunshine.shape}, dtype: {sunshine.dtype}")
```

```python
sunshine_gray = imread_gray('sunshine.jpg')
plt.imshow(sunshine_gray, cmap='gray')
```

```python
figure, axis = plt.subplots(1, 3, figsize=(20,5))
percentiles = [0.25, 0.5, 0.75]
for percent, ax in zip(percentiles, axis):
    thresh_value = int(percent * 255)
    _, sunshine_binary = cv2.threshold(sunshine_gray, thresh_value, 255, cv2.THRESH_BINARY)
    
    ax.imshow(sunshine_binary, cmap='gray')
    ax.set_title(f'Threshold at {thresh_value} ({percent*100})')
    ax.axis('off')
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20,5))
ax[0].imshow(sunshine)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(sunshine[:,:,0], cmap='Reds')
ax[1].set_title('Red Channel')
ax[1].axis('off')
ax[2].imshow(sunshine[:,:,1], cmap='Greens')
ax[2].set_title('Green Channel')
ax[2].axis('off')
ax[3].imshow(sunshine[:,:,2], cmap='Blues')
ax[3].set_title('Blue Channel')
ax[3].axis('off')
```

```python
sunshine_hsv = cv2.cvtColor(sunshine, cv2.COLOR_RGB2HSV)
fig, ax = plt.subplots(1, 4, figsize=(20,5))
ax[0].imshow(sunshine)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(sunshine_hsv[:,:,0], cmap='hsv')
ax[1].set_title('Hue Channel')
ax[1].axis('off')
ax[2].imshow(sunshine_hsv[:,:,1], cmap='hsv')
ax[2].set_title('Saturation Channel')
ax[2].axis('off')
ax[3].imshow(sunshine_hsv[:,:,2], cmap='hsv')
ax[3].set_title('Value Channel')
ax[3].axis('off')
```
